1. What is the difference between R-square and adjusted R square and why is it desirable to use the adjusted value? (4 pts)

2. Overfitting usually happens in complex models. Linear Regression is a fairly simple model. Could overfitting happen in Linear Regression? If so, please explain the scenario in which it could happen and how we can tackle it. (6 pts)


1. 
Both R-square and adjusted R-suqare are statistical measures used to determine the goodness of fit of a regresion     model. Both these statistics capture how well the predictors in a regression model explains the variability in the     traget/dependent variable.

The difference between R-square and adjusted R square is that R-square does not take regression model complexity into  consideration. R-squared may falsely suggest that adding more predictors is always better, even if they don't add meaningful explanatory power. It always increases as more independent variables are added to the model, even if they do not contribute significantly in explaining the dependent variable potentially resulting in overfitting.

Adjusted R-square allows to assess whether adding more predictors is justified in terms of improving model performance. Adjusted R-square takes model complexity into considertaion and penalizes the inclusion of unnecessary variables that do not significantly contribute to explaining the dependent variable. If added variables do not contribute significantly, adjusted R2 will decrease or remain the same.

When computing the value of R-squared, the adjusted R-squared takes into consideration the number of predictors. It is  more appropriate to use adjusted R-square than R-squared because it provides a more accurate measure for the fit of the model by taking model complexity into consideration, penalizing unnecessary variables and thus preventing over-fitting.


2. YES, overfitting can occur in Linear Regression models. 

*Reasons and how to tackle:*

(I) Overfitting can occur in Linear Regression models if the size of the training data is too small. Due to the limited amount of data, the model may overfit the training data by fitting it too closely. As a result, it performs well on the training data but fails to generalize and hence perform poorly on test data.

- One way to tackle this situation is to increase the size of the dataset. Another approach is to use cross validation techniques to increase the model's generalization capacity.


(II) Having an excessive quantity of predictive features in relation to the number of observations can also lead to overfitting. Adding a large number of potential predictors to a small dataset prompts the model to mold very closely to peculiarities in that limited data. This may limit generalizability.
  
- Regularization techniques such as Ridge or Lasso regression can be beneficial as they reduces the model complexity by coefficient shrinkage.

## Part-2: *(5 pts)*

Use `sklearn.linear_model.Lasso` and `sklearn.linear_model.Ridge` classes to do a 5-fold cross validation using sklearn's `KFold`. For the sweep of the regularization parameter, we will look at a grid of values ranging from  $\alpha=10^{-6}$  to  $\alpha=10^{6}$.In Python, you can consider this range of values as follows: `alpha = 10**numpy.linspace(-6, 6, 100)` so that you can generate 100 uniform values between -6 to 6 as power series.

Fit the 2 regression models (Lasso and Ridge) with scaled data and report the best chosen $\alpha$ based on cross validation as well as the corresponding scoring metric. The cross validation should happen on your training data using MSE as the scoring metric.

```python
# Define number of folds
##  START CODE  ## (1 line of code)
n_folds = 5
##  END CODE  ##
```

```python
# Create KFold from sklearn
##  START CODE  ## (1 line of code)
k_fold =KFold(n_folds, random_state=50, shuffle=True)
##  END CODE    ##
```

```python
#Define the alphas as defined in the question
##  START CODE  ## (1 line of code)
alphas =10**np.linspace(-6, 6, 100)
##  END CODE    ##
lasso_avg_mse = {}
```

```python
#For each value of alpha and each fold compute the mean square error
for alpha in alphas:

  #Instantiate a lasso model with the current alpha
  ##  START CODE  ## (1 line of code)
  lasso = linear_model.Lasso(alpha=alpha)
  ##  END CODE    ##

  avg_mse = 0

  for k, (train, test) in enumerate(k_fold.split(Xscaled_train, y_train)):

    #Fit the scaled training data to the lasso model
    ## START CODE ## (1 line of code)
        lasso.fit(Xscaled_train[train],y_train[train])
    ## END CODE ##

    #Calculate the average mean sqaured error
    ##  START CODE  ## (1 line of code)
        avg_mse += mean_squared_error(y_train[test],lasso.predict(Xscaled_train[test]))
    ##  END CODE    ##

  # Take the average mean squared error as metric
  lasso_avg_mse[alpha] = avg_mse / n_folds
```

```python
# Find the best value for alpha with minimum mean squared error
##  START CODE  ## (1 line of code)
best_alpha_lasso = pd.Series(lasso_avg_mse).idxmin()
##  END CODE    ##
print("Best lasso alpha: {}".format(best_alpha_lasso))
```

```python
#For each value of alpha and each fold compute the mean square error
ridge_avg_mse = {}
for alpha in alphas:

  #Instantiate a ridge model with the current alpha
  ##  START CODE  ## (1 line of code)
  ridge = linear_model.Ridge(alpha=alpha)
  ##  END CODE    ##

  avg_mse = 0

  for k, (train, test) in enumerate(k_fold.split(Xscaled_train, y_train)):

    #Fit the scaled training data to the ridge model
    ## START CODE ## (1 line of code)
      ridge.fit(Xscaled_train[train], y_train[train]) 
    ## END CODE ##

    #Calculate the average mean sqaured error
    ##  START CODE  ## (1 line of code)
      avg_mse += mean_squared_error(y_train[test],ridge.predict(Xscaled_train[test])) 
    ##  END CODE    ##

  # Take the average mean squared error as metric
  ridge_avg_mse[alpha] = avg_mse / n_folds
```

```python
# Find the best value for alpha with minimum mean squared error
##  START CODE  ## (1 line of code)
best_alpha_ridge =pd.Series(ridge_avg_mse).idxmin()
##  END CODE    ##

print("Best Ridge alpha: {}".format(best_alpha_ridge))
```

Why is it advisable to scale the independent variables when applying ridge or  lasso regression?

Ridge and Lasso regression techniques add penalty terms to the linear regression cost function. These penalty terms are multiplied by the coefficients of the independent variables. When the independent variables have different scales, the coefficients can be penalized unfairly.

By scaling these features, Ridge and Lasso regression treat all of them equally in terms of the regularization penalty. This ensures that the model doesn't give undue importance to one feature over another based solely on its scale, leading to a more fair and reliable prediction model
